Packages allow python code to logically ordered and reused.

If you often find yourself writing the same or similar bits of code, it might be an indicaiton you should package it up for easier reusability.

`Pip` is the package manager for python.

To install a package with pip, use: `pip install <name_of_package>`.

You can also view installed packages with `pip list` or look for a specific package with `pip show <name_of_package>`.

Some packages are installed by default with python, such as `os` and `math`.

Additionaly, you can create and use your own python code by adding it to your path, or saving it to your working directory.

Be cautious of naming your files anything that is already in use. For example if you create a math.py file in your local directory, it will block you from importing that actual math package in python (demonstrated below). 

# Explore OS Library

```python
# Import the libary
import os
```

```python
# Use dir to see all methods and functions
dir(os)
```

```python
# Use os.getcwd to create a variable for our current working location
## Notice that `os.` is required at the beginning so python can find the appropriate function
home = os.getcwd()
print(home)
```

```python
# See where the package is installed
package_location = os.__file__
print(package_location)
```

```python
# Check out what else is installed in that directory
dir_path = os.path.dirname(package_location)
print(dir_path)
```

```python
# change to that directory and list contents
os.chdir(dir_path)
```

# Custom Code Imports

The python file `MyFirstImport.py` in this directory can be imported just like a package.

When naming your custom code, try to avoid existing names of packages such as math, os or pandas. Use of those names can cause collisions and result in unexpected behavior. 

```python
import MyFirstImport as MFI
```

```python
dir(MFI)
```

```python
MFI.SayHello()
```

```python
MFI.SayHello?
```

```python
MFI.SayHello('Danny')
```

# Understanding Import Hierarchy

The `path` for python shows where it can import from and in which order.

The path starts with the current working directory and then looks recursively through the path to find a matching import.

If it can not find one, it will report `module not found`.
