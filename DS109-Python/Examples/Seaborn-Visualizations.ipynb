```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.shape
```

```python
mpg.head()
```

```python
mpg.origin.unique()
```

```python
sns.displot(data=mpg, x='mpg')
```

```python
sns.scatterplot(data=mpg, x='mpg', y='horsepower', hue='origin')
```

```python
sns.scatterplot(data=mpg, x='mpg', y='horsepower', hue='model_year')
```

```python
sns.violinplot(data=mpg, x='origin', y='mpg')
```

```python
sns.violinplot(data=mpg, x='origin', y='mpg', hue='cylinders')
```

```python
sns.scatterplot(data=mpg, x='acceleration', y='weight')
```

```python
sns.scatterplot(data=mpg, x='acceleration', y='weight', hue='origin')
```

```python
sns.scatterplot(data=mpg, x='acceleration', y='weight', hue='model_year')
```

```python
sns.barplot(data=mpg, x='model_year', y='weight')
```

```python
sns.barplot(data=mpg, x='model_year', y='mpg')
```

```python
sns.barplot(data=mpg, x='origin', y='mpg')
```

```python
sns.barplot(data=mpg, x='origin', y='mpg', palette="Purples")
```

```python
numeric_only = mpg[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight','acceleration']]
```

```python
mpg.columns
```

```python
sns.heatmap(numeric_only.corr(), cmap='Purples')
```

```python
sns.heatmap(numeric_only.corr(), cmap='Blues')
```

```python
sns.barplot(data=mpg, x='model_year', y='mpg').set_title("My Favorite Bar Graph")
```

```python
g = sns.barplot(data=mpg, x='model_year', y='mpg')
g.set_title("My Favorite Bar Graph")
g.set_xticklabels(g.get_xticklabels(), rotation=30)
```

```python
g.figure.savefig("output.png")
```
